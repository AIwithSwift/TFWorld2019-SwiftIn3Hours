```
# Functions and optionals

def scaled(collection, multiplier):
    new_collection = []

    for element in collection:
        value = element if element is not None else 0
        new_collection.append(value * multiplier)

    return new_collection

values = [1, None, 6, 4, 8, 2]
print(scaled(values, 10))
```

```
# Tuples, dataclasses and classes
tuplet = (True, False)
tuplet = (False, False)

from typing import NamedTuple

class Dog(NamedTuple):
    name: str
    color: str
    gender: str
    age: int

myDog = Dog("Argos", "Whippet", "Male", 7)

# Python 3.7+
from dataclasses import dataclass

@dataclass
class NewDog:
    name: str
    color: str
    gender: str
    age: int

myDog = NewDog("Argos", "Whippet", "Male", 7)

class Cat:
    def __init__(self, name, color, gender, age):
        self.name = name
        self.color = color
        self.gender = gender
        self.age = age

    def have_birthday(self):
        self.age += 1

myCat = Cat("Lola", "Gray", "Female", 5)
myCat.have_birthday()
print(myCat.age)
```

```
# Enumerations

from enum import Enum

class Day(Enum):
    SUNDAY = 0
    MONDAY = 1
    TUESDAY = 2
    WEDNESDAY = 3
    THURSDAY = 4
    FRIDAY = 5
    SATURDAY = 6
    
print("MONDAY is day", Day.MONDAY.value)
print(Day(3).name, "is day 3")
```

```
# define a class called '' with the following properties
# name - a string
# days - an integer
class Conference:
    def __init__(self, name, days):
        self.name = name
        self.days = days

    def describe(self):
        print("A conference called", self.name, "that runs for", str(self.days), "days.")

# initialise an instance for TensorFlow World with a length of 4 days
tfw = Conference("TensorFlow World", days=4)

tfw.describe()
```

```
# Cleaning Data
collection = [0, 6, 2, 5, 5, 1, 8, 10]

# lambdas!
doubles = list(map(lambda x: x * 2, collection))
print(doubles)

evens = list(filter(lambda x: x % 2 == 0, collection))
print(evens)

total = 0
for element in collection:
    total += element
print(total)

# function passing
def nonsense_function(number):
    return number + 4

def apply_sporadic(collection, stride, function):
    new_collection = collection

    for index, element in enumerate(collection):
        if index % stride == 0:
            new_collection[index] = function(element)

    return collection

thing = [1, 1, 1, 1, 1, 1, 1, 1]
thing = apply_sporadic(thing, 3, nonsense_function)
print(thing)
```
