# Extra - Raw TensorFlow Operators

This extra is based on the Swift for TensorFlow Raw TensorFlow operators notebook. It is updated.

## Imports

```swift
import TensorFlow
```

Use [`Raw`](https://www.tensorflow.org/swift/api_docs/Enums/Raw) to call raw operations:

```swift
Raw.mul(Tensor([2.0, 3.0]), Tensor([5.0, 6.0]))
```

```swift
infix operator • : MultiplicationPrecedence
```

```swift
extension Tensor where Scalar: Numeric {
    static func • (_ lhs: Tensor, _ rhs: Tensor) -> Tensor {
        return Raw.mul(lhs, rhs)
    }
}
```

```swift
let x: Tensor<Double> = [[1.0, 2.0], [3.0, 4.0]]
let y: Tensor<Double> = [[8.0, 7.0], [6.0, 5.0]]
x • y
```
